```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("natural_disasters_data.csv")
df.head()
```

```python
df_no_pand_no_famine = df[0:10]
df_famine = df[10:15]
df_pand = df[15:20]
```

```python
def plot_data(df, title, xlabel, ylabel, filename, rotation=60):
    x = np.array(df["Event"])
    y = np.array(df["Death Toll"])
    plt.style.use('ggplot')
    plt.bar(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(rotation=rotation)
    plt.savefig(filename, bbox_inches='tight')
```

```python
plot_data(df_no_pand_no_famine, "Death toll by Event (no famines, epidemics, or pandemics)", "Event", "Death toll by Event (no famines, epidemics, or pandemics)", "nopandnofam.png")
```

```python
plot_data(df_famine, "Death toll by Famine", "Famine", "Death toll by Famine (10 million)", "fam.png")
```

```python
plot_data(df_pand, "Death toll by Disease", "Disease", "Death toll by Disease (1 billion)", "dis.png")
```

```python
plot_data(df, "Death toll by Occurrence", "Occurrence", "Death toll by Occurrence (1 billion)", "all.png", rotation=90)
```
